```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_Venezuela = pd.read_csv("/content/drive/MyDrive/Project 1 - DS 4002/Sentiment Scores/Venezuela Sentiment.csv")

print(df_Venezuela)
```

```python
#Venezuela Heat Map
heat = df_Venezuela.set_index('News Organization')[['Negative','Neutral','Positive']]
sns.heatmap(heat, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Sentiment Heatmap')
plt.show()
```

```python
#Venezuela Bar Graph
df_melt = df_Venezuela.melt(
    id_vars='News Organization',
    value_vars=['Negative','Neutral','Positive'],
    var_name='Sentiment',
    value_name='Score'
)

plt.figure(figsize=(12,6))
sns.barplot(
    data=df_melt,
    x='News Organization',
    y='Score',
    hue='Sentiment'
)

plt.xticks(rotation=45, ha='right')
plt.title('Sentiment by News Organization')
plt.legend(title='Sentiment', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.show()
```

```python
#Venezuela Net Sentiment Graph
df_Venezuela['Net Sentiment'] = df_Venezuela['Positive'] - df_Venezuela['Negative']
df_Venezuela.sort_values('Net Sentiment', ascending=False)

df_Venezuela.sort_values('Net Sentiment').plot(
    x='News Organization',
    y='Net Sentiment',
    kind='barh'
)
plt.axvline(0)
plt.title('Net Sentiment by News Organization')
plt.show()
```
